- Q1. Write a program to retrieve the keys/values of dictionary

**Use the dictionary**

dictionary2 = {0:3, 'x':5, 1:2}

```python
dictionary2 = {0:3, 'x':5, 1:2}
for x in dictionary2:
    
 #   print(x)
    print("key is: {0}, value is:{1}".format(x,dictionary2[x]))
          
          
          
          
```

```python
dictionary2 = {0:3, 'x':5, 1:2}
print(dictionary2.items())
          
items=tuple(dictionary2.items())
print(items)
          
for i in items:
    key,value =i
    print("key is: "+str(key)+"value is: "+str(value))
```
