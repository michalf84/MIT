```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df
```

```python
df.drop("Unnamed: 0", axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
sales=df["Sales"].values.reshape(len(df),1)
tv=df["TV"].values.reshape(len(df),1)
radio=df["Radio"].values.reshape(len(df),1)
newspapers=df["Newspaper"].values.reshape(len(df),1)
```

```python
tv_model=linear_model.LinearRegression()
tv_model.fit(tv,sales)
```

```python
coef_tv=np.array(list(tv_model.intercept_.flatten())+list(tv_model.coef_.flatten()))
```

```python
coef_tv=list(coef_tv)
coef_tv
```

```python
newspaper_model=linear_model.LinearRegression()
newspaper_model.fit(newspapers,sales)
```

```python
coef_news=np.array(list(newspaper_model.intercept_.flatten())+list(newspaper_model.coef_.flatten()))
coef_news=list(coef_news)
coef_news
```

```python
radio_model=linear_model.LinearRegression()
radio_model.fit(radio,sales)
```

```python
coef_radio=np.array(list(radio_model.intercept_.flatten())+list(radio_model.coef_.flatten()))
coef_radio=list(coef_radio)
coef_radio
```

```python
dict_sales={}
```

```python
dict_sales["tv"]=coef_tv
dict_sales
```

```python
dict_sales["radio"]=coef_radio
dict_sales
```

```python
dict_sales["news"]=coef_news
dict_sales
```

```python
metric=pd.DataFrame(dict_sales)
```

```python
metric.index=["interc","coef"]
metric
```

```python
metric.loc['r-squared']=[tv_model.score(tv,sales),radio_model.score(radio,sales),newspaper_model.score(newspapers,sales)]
```
